# Exploratory Data Analysis - Credit Card Fraud Detection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

print('Import successful!')
```

```python
df = pd.read_csv('./data/creditcard.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
missing = df.isnull().sum()

if missing.sum() == 0:
    print("No missing values in the dataset.")
else:
    print("Missing values detected:")
    print(missing[missing > 0])
```

```python
df.describe()
```

```python
max_time_hours = df['Time'].max() / 3600
print(f"Maximum transaction time: {max_time_hours:.2f} hours, days: {max_time_hours/24:.2f} days")
```

```python
counts = df['Class'].value_counts()
legit = (counts[0] / len(df)) * 100
fraud = (counts[1] / len(df)) * 100

legit, fraud
```

```python
plt.figure(figsize=(8, 5))
counts.plot(kind='bar', color=['green', 'red'])
plt.title('Class Distribution (Legitimate vs Fraud)', fontsize=14, fontweight='bold')
plt.xlabel('Class (0 = Legit, 1 = Fraud)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)

# Add count labels on bars
for i, v in enumerate(counts):
    plt.text(i, v + 5000, f'{v:,}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
print(f'Percentage: 1:{int(counts[0]/counts[1])}')
```

For every fraud, we have about 577 legit transactions. Standard machine learning metrics cannot work for this problem because of it's high imbalance. An average model without any feature engineering should be 99.83% accurate - therefore, proving that Accuracy is not the best metric to use here.

```python
df.groupby('Class')['Amount'].quantile(
    [0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]
).unstack()
```

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.yscale('log')
plt.title('Transaction Amount by Class (Log Scale)')
plt.show()
```

Fraudulent transactions tend to involve lower amounts than legitimate transactions, with a median value of 9.25 compared to 22.00. However, fraud also exhibits a heavier upper tail, indicating the presence of occasional high-value fraudulent transactions. This suggests that transaction amount alone is insufficient for fraud detection and must be combined with other behavioral features.
