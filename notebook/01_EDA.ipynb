# Exploratory Data Analysis — Credit Card Fraud Detection

## Business Objective
The goal is to build a fraud detection system that maximizes business profit by:
- Minimizing financial loss from missed fraud (false negatives)
- Minimizing customer friction from false positives
- Operating under real-world class imbalance (0.17% fraud)

This notebook focuses on understanding the data, identifying risk patterns,
and defining metrics that will guide model development.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/creditcard.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
fraud_count = df['Class'].value_counts()
percent = fraud_count / len(df) * 100
print(fraud_count)
print(percent)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Class distribution')
plt.savefig('../output/eda/class_distribution.png')
plt.show
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
```

```python
# Fraud vs Amount
plt.figure(figsize=(7,6))
sns.boxplot(x='Class', y='Amount', data=df)
plt.yscale('log')
plt.title('Transaction Amount by Class')
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.countplot(x='Hour', hue='Class', data=df)
plt.title("Fraud by Hour of Day")
plt.show()
```

```python
for col in df.columns[:5]:
    plt.figure(figsize=(6,4))
    sns.kdeplot(df[df['Class']==0][col], label="Legit")
    sns.kdeplot(df[df['Class']==1][col], label="Fraud")
    plt.title(f"Distribution of {col}")
    plt.legend()
    plt.show()
```

## Key EDA Findings

- The dataset is extremely imbalanced (~0.17% fraud), making accuracy unusable; evaluation must focus on recall, precision, PR-AUC, and business profit.
- Transaction patterns vary by time of day, suggesting temporal features (e.g., hour-of-day, transaction velocity) may improve fraud detection.
- Several anonymized features (V1–V28) exhibit distributional shifts between fraud and legitimate transactions, supporting the use of non-linear, interaction-aware models.
- The asymmetric cost of errors requires cost-sensitive learning and decision-threshold optimization to balance fraud capture and customer friction.
